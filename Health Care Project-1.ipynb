## Project Task: Week 3
**1) Devise strategies for model building. It is important to decide the right validation framework. Express your thought process

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
X = data.drop(["Outcome"],axis=1)
Y = data["Outcome"]
```

```python
X.isna().sum()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
model = SVC(kernel='linear',C=0.005958,gamma='auto').fit(X_train,Y_train)
```

```python
predict = model.predict(X_test)
```

```python
accuracy_score(Y_test,predict)
```

```python
model1 = LogisticRegression(C=0.0005,max_iter=50000,n_jobs=5).fit(X_train,Y_train)
```

```python
predict_1 = model1.predict(X_test)
```

```python
accuracy_score(Y_test,predict_1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_2 = DecisionTreeClassifier(max_depth=5,ccp_alpha=0.00575).fit(X_train,Y_train)
```

```python
predict_2 = model_2.predict(X_test)
```

```python
accuracy_score(Y_test,predict_2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_3 = RandomForestClassifier(n_estimators=20000).fit(X_train,Y_train)
```

```python
predict_3 = model_3.predict(X_test)
```

```python
accuracy_score(Y_test,predict_3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_4 = GaussianNB().fit(X_train,Y_train)
```

```python
predict_4 = model_4.predict(X_test)
```

```python
accuracy_score(Y_test,predict_4)
```

```python
from xgboost import XGBClassifier
```

```python
model_5 = XGBClassifier(n_estimators=10000,learning_rate=0.007).fit(X_train,Y_train)
```

```python
predict_5 = model_5.predict(X_test)
```

```python
accuracy_score(Y_test,predict_5)
```
